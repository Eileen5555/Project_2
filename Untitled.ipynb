```python
##############################################DAY 3 WEEK 5 ###################################################################
#merge data frames 
df_merged = pd.merge(df_web_data, df_clients, on='client_id')

# Convert date_time to datetime
df_merged['date_time'] = pd.to_datetime(df_merged['date_time'])

# Sort values by client_id and date_time
df_merged = df_merged.sort_values(by=['client_id', 'date_time'])

# Calculate time differences
df_merged['time_diff'] = df_merged.groupby('client_id')['date_time'].diff()

# Identify previous step
df_merged['prev_step'] = df_merged.groupby('client_id')['process_step'].shift(1)


def calculate_completion_rate(df):
    total_users = df['client_id'].nunique()
    completed_users = df[df['process_step'] == 'confirm']['client_id'].nunique()
    return completed_users / total_users * 100

def calculate_avg_time_per_step(df):
    return df.groupby('process_step')['time_diff'].mean()

def calculate_error_rate(df):
    df['error'] = df['process_step'] < df['prev_step']
    return df['error'].mean() * 100

def calculate_drop_off_rates(df, total_users):
    drop_off_counts = df.groupby('process_step')['client_id'].nunique()
    return drop_off_counts / total_users * 100

def calculate_bounce_rate(df):
    first_steps = df.groupby('client_id').first().reset_index()
    return (first_steps['process_step'] != 'confirm').mean() * 100
```

```python
# Separate control and test groups
df_control = df_merged[df_merged['Variation'] == 'Control']
df_test = df_merged[df_merged['Variation'] == 'Test']

# Control group KPIs
control_completion_rate = calculate_completion_rate(df_control)
control_avg_time_per_step = calculate_avg_time_per_step(df_control)
control_error_rate = calculate_error_rate(df_control)
control_drop_off_rates = calculate_drop_off_rates(df_control, df_control['client_id'].nunique())
control_bounce_rate = calculate_bounce_rate(df_control)

# Test group KPIs
test_completion_rate = calculate_completion_rate(df_test)
test_avg_time_per_step = calculate_avg_time_per_step(df_test)
test_error_rate = calculate_error_rate(df_test)
test_drop_off_rates = calculate_drop_off_rates(df_test, df_test['client_id'].nunique())
test_bounce_rate = calculate_bounce_rate(df_test)

# Display results
print(f"Control Group - Completion Rate: {control_completion_rate:.2f}%")
print(f"Control Group - Average Time per Step:\n{control_avg_time_per_step}")
print(f"Control Group - Error Rate: {control_error_rate:.2f}%")
print(f"Control Group - Drop-off Rates:\n{control_drop_off_rates}")
print(f"Control Group - Bounce Rate: {control_bounce_rate:.2f}%")

print(f"\nTest Group - Completion Rate: {test_completion_rate:.2f}%")
print(f"Test Group - Average Time per Step:\n{test_avg_time_per_step}")
print(f"Test Group - Error Rate: {test_error_rate:.2f}%")
print(f"Test Group - Drop-off Rates:\n{test_drop_off_rates}")
print(f"Test Group - Bounce Rate: {test_bounce_rate:.2f}%")
```

```python
# Calculate completion rates
control_completion_rate = df_control_merged[df_control_merged['process_step'] == 'confirm']['client_id'].nunique() / df_control_merged['client_id'].nunique()
test_completion_rate = df_test_merged[df_test_merged['process_step'] == 'confirm']['client_id'].nunique() / df_test_merged['client_id'].nunique()

# Calculate sample sizes and successes
n_control = df_control_merged['client_id'].nunique()
n_test = df_test_merged['client_id'].nunique()
successes_control = df_control_merged[df_control_merged['process_step'] == 'confirm']['client_id'].nunique()
successes_test = df_test_merged[df_test_merged['process_step'] == 'confirm']['client_id'].nunique()

# Two-proportion z-test for completion rates
count = np.array([successes_control, successes_test])
nobs = np.array([n_control, n_test])
stat, pval = proportions_ztest(count, nobs)
print(f"Z-test for completion rates: p-value = {pval:.4f}")

# One-proportion z-test for cost-effectiveness threshold
effect_size = 0.05
pooled_success_rate = (successes_control + successes_test) / (n_control + n_test)
standard_error = np.sqrt(pooled_success_rate * (1 - pooled_success_rate) * (1/n_control + 1/n_test))
test_statistic = (test_completion_rate - control_completion_rate - effect_size) / standard_error
pval_threshold = 1 - proportions_ztest([successes_control, successes_test], [n_control, n_test])[1]
print(f"One-proportion z-test for cost-effectiveness threshold: p-value = {pval_threshold:.4f}")
```
